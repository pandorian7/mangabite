```python
import requests
import bs4
from urllib.parse import urlparse, urlunparse
from time import sleep
```

```python
url = 'https://www.webtoons.com/en/mystery/senorita-cometa/list?title_no=3387'

def get(url):
    res = requests.get(url)
    print(url, res.status_code)
    return res.content

def build_url(base, path):
    return urlunparse(urlparse(base)._replace(**{key:val for key, val in urlparse(path)._asdict().items() if val}))

def get_series_pages(url):

    data = get(url)
    soup = bs4.BeautifulSoup(data, 'html.parser')
    elements = soup.select('div.paginate a')
    links = map(lambda link:url if link == '#' else build_url(url, link), [ele['href'] for ele in elements])
    return list(links)[::-1]

def get_chapters_in_page(page_url):
    data = get(page_url)
    soup = bs4.BeautifulSoup(data, 'html.parser')
    chapter_links = [link['href'] for link in soup.select('div.detail_lst ul li a')][::-1]
    return chapter_links
    
def get_chapter_links(url):
    pages = get_series_pages(url)
    for page in pages:
        chapter_links = get_chapters_in_page(page)
        for chapter_link in chapter_links:
            yield chapter_link

for chap in get_chapter_links(url):
    print(chap)
```
